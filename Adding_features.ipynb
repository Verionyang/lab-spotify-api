```python
def search_song(title,artist):
    query=title+artist
    try: 
        string = "tracks: " + title + "artist: " + artist
        song=sp.search(q=string,limit=1)
        id = song['tracks']['items'][0]['id']
    except:
        #print("Song not found!")
        id = np.nan
    
    return id
```
